## Question 1

```python
#question 1
k4=[] #create empty list to append the palindrome value
for i in range(1,1000): #first variable from range 1 to 999
  for j in range(1,1000): #second variable from range 1 to 999
    k1=i*j #the product of variable 1 and variable 2 and loop to get all the posible product 
    k2 = str(k1) #change it to string
    k3 = k2[::-1] #make a reverse string
    if k2==k3: #condition for palindrome
      k4.append(k1) 

max(k4) #find the highest palindrome
```

for this question, we used the 2 times for loop approached to get the all posible multiple of the two variable. then we added all the value that matched the condition into one array to find its max value.

## Question 4

```python
# question 4
def compresStr(words): #create a function for compression string
    compresStr_words = "" # string variable to insert the sum char
    i = 0 #to start iteration from index 0
   
    while (i <= len(words)-1): #the loop for every words
        count = 1 #starts the count from 1
        charac_in_word = words[i] #if same then insert to variable charac_in_word
        j = i
        while (j < len(words)-1): # loop again to find how many char are in that words
            if (words[j] == words[j+1]): #if many, than into the condition and sum up
                count = count+1
                j = j+1
            else:
                break
        compresStr_words=compresStr_words+ charac_in_word + str(count) #then total up from the loop into one variable
        i = j+1
    return compresStr_words #return the compression string
  
compresStr_words=compresStr("AAABBBDDDCVVvvvVV") #use the function above
print(compresStr_words) #print the result
```

For this question, we created a function for it. Then inside the function there are a while loop as it is easier for this type of condition comparing with a for loop. so if the condition match, it will sum the num of words inside the string and printed the total result as shown and explained above.
